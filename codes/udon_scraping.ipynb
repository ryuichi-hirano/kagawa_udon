```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import time
from xml.etree.ElementTree import parse
import urllib.parse
import urllib.request
import googlemaps
import json
```

```python
res = requests.get('https://www.my-kagawa.jp/udon/list')
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
link_l = []

num = 1
while True:
    try:
        res = requests.get('https://www.my-kagawa.jp/udon/list/page:'+str(num))
    except:
        break

    soup = BeautifulSoup(res.text, 'html.parser')
    soup_link = soup.find_all("h3", class_="pull-left")
    # print(soup_link)

    if not soup_link:
        break

    for sl in soup_link:
        # print(sl.a.get('href'))
        link_l.append([sl.text.replace('\u3000', ' '), sl.a.get('href')])

    num += 1
```

```python
for l in link_l:
    try:
        res = requests.get('https://www.my-kagawa.jp'+str(l[1]))
    except:
        continue

    soup = BeautifulSoup(res.text, 'html.parser')
    shop_base = soup.find("table", class_="base1")
    try:
        shop_address = shop_base.td.text.split('　')[1]
    except:
        shop_address = shop_base.td.text
    print(shop_address)
    l.append(shop_address)
    
```

```python
shop_l = link_l
```
